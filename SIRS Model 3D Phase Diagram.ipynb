#  SIRS MODEL 3D PHASE PORTRAIT
This is a little hard to see, but you can mess around with the 3D phase portrait as well.  Here all vectors are normalized to have the same length, so the magnitude of the vectors will not tell you how quickly the system is changing at that point.

```python
beta=.3
nu=1
gamma=.3

Sinitial=100
Iinitial=2
Rinitial=0

N=Sinitial + Iinitial + Rinitial

ReproductiveNumber=N*beta/nu

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
fig = plt.figure()
ax = fig.gca(projection='3d')
x, y, z = np.meshgrid(np.arange(0, 100, 10),
                      np.arange(0, 100, 10),
                      np.arange(0, 100, 10))
u = -beta*x*y+gamma*z
v = beta*x*y-nu*y
w = nu*y-gamma*z
print('The value of N*beta/nu =', ReproductiveNumber, ', and there will be an epidemic if this number is greater than 1')
ax.quiver(x, y, z, u, v, w,length=8,normalize=True)
plt.xlabel('S')
plt.ylabel('I')
plt.show()
```
